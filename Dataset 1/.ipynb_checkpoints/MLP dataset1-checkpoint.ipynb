```python
torch.set_default_dtype(torch.float64)

from torch import nn, optim
import torch.nn.functional as F

class My_module(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc0 = nn.Linear(19,36)
        self.fc00 = nn.Linear(32,64)
        
        self.conv1=nn.Conv2d(1,8,kernel_size=(3,3),stride=(1,1),padding=1)
        self.conv2=nn.Conv2d(8,12,kernel_size=(3,3),stride=(1,1),padding=1)
        self.conv3=nn.Conv2d(12,18,kernel_size=(3,3),stride=(1,1),padding=1)
        self.fc1 = nn.Linear(18*6*6,350)
        self.fc2 = nn.Linear(350,4)
        
        self.fc3 = nn.Linear(56, 64)
        self.fc4 = nn.Linear(64, 4)
        self.fc5 = nn.Linear(400, 4)
        self.dropout = nn.Dropout(p=0.0)
        self.pool=nn.MaxPool2d(1,1)
        
    def forward(self, x):
        # make sure input tensor is flattened
        #x = x.view(x.shape[0], -1)
        #print(x.shape)
        x = self.dropout(self.fc0(x))
       # x = self.dropout(F.relu(self.fc00(x)))
        #print(x.shape)
        x=x.view(-1,1,6,6)
        x=self.pool(F.relu(self.conv1(x)))
        #print(x.shape)
        x=self.pool(F.relu(self.conv2(x)))
        #print(x.shape)
        x=self.pool(F.relu(self.conv3(x)))
        #print(x.shape)
        
        x=x.view(-1,18*6*6)
        #print(x.shape)
        x = self.dropout(F.relu(self.fc1(x)))
        #print(x.shape)
        
 #       x = self.dropout(F.relu(self.fc2(x)))
        #x = self.dropout(F.relu(self.fc3(x))) 
        #x = self.dropout(F.relu(self.fc4(x))) 
    
        x = F.log_softmax(self.fc2(x), dim=1)
        #print(x.shape)
        
        return x
    
model = My_module()
    
#model = nn.Sequential(nn.Linear(11, 28),
 #                     #nn.GELU(),
  #                    #nn.ELU(),
   #                   nn.ReLU(),
    #                  nn.Dropout(p=0.3),                      
     #                 #nn.LeakyReLU(),
      #                #nn.Linear(64, 140),
       #               #nn.ReLU(),
                      #nn.Linear(140, 200),
                      #nn.ReLU(),
        #              nn.Linear(28,32),                      
                      #nn.LeakyReLU(),
                      #nn.ELU(),
                      #nn.GELU(),
         #             nn.ReLU(),
          #            nn.Dropout(p=0.3),
           #           nn.Linear(32, 40),
            #          nn.ReLU(),
             #         nn.Dropout(p=0.3),
       #               nn.Linear(40, 56),
        #              nn.ReLU(),
         #             nn.Dropout(p=0.3),
          #            nn.Linear(56, 4),
           #           nn.LogSoftmax(dim=1)
                      #nn.LogSoftmax(dim=1)
            #         )

#criterion = nn.NLLLoss()
Fonction_de_perte = nn.NLLLoss()

optimizer = optim.SGD(model.parameters(), lr=0.003)

#indices = [0,3, 299:303]
```

```python

torch.set_default_dtype(torch.float64)

model = nn.Sequential(nn.Linear(11, 64),
                      nn.ReLU(),
                      nn.Linear(64, 140),
                      nn.ReLU(),
                      nn.Linear(140, 200),
                      nn.ReLU(),
                      nn.Linear(200, 32),
                      nn.ReLU(),
                      nn.Linear(32, 4),
                      nn.LogSoftmax(dim=1))

#criterion = nn.NLLLoss()
Fonction_de_perte = nn.CrossEntropyLoss()

optimizer = optim.SGD(model.parameters(), lr=0.003)

#indices = [0,3, 299:303]
```

```python
#labels1=labels[loc]

inputs =torch_tensor[:, :] 
labels1=labels[:]


epochs = 10
#print(inputs.shape[0])

for e in range(epochs):
    running_loss = 0
    for i, x in enumerate(inputs):
        optimizer.zero_grad()
        x2=x[None,:]
        output = model.forward(x2)
        l2=labels1[i][None]
        #print(l2)
        loss = Fonction_de_perte(output, l2)
      
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item()
    
    else:
        #print(l2)
        print(f"Training loss: {running_loss/len(inputs)}")
```
